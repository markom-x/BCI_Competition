```python
import numpy as np
import matplotlib.pyplot as plt
import mne
from mne.preprocessing import ICA

raw = mne.io.read_raw_gdf("A01T.gdf", preload=True)
raw.filter(1., 40.).notch_filter(50.)
raw.set_channel_types({'EOG-left':'eog',
                       'EOG-central':'eog',
                       'EOG-right':'eog'})
```

```python
raw.plot(picks=['EEG-Fz', 'EOG-left', 'EOG-central', 'EOG-right'],
         scalings={'eeg':30}, n_channels=4, block=True,
         duration=10.0,
         title='Fz vs EOG: trova i blink')
```

```python
# --- (1) ICA per blink -------------------------------------------------
ica = ICA(n_components=20, random_state=97, max_iter='auto')
ica.fit(raw.copy().pick_types(eeg=True, eog=True))
eog_inds, _ = ica.find_bads_eog(raw, threshold=3.0)
ica.exclude = eog_inds
clean_raw = raw.copy()
ica.apply(clean_raw)

# --- (2) estrai Fz pulito ---------------------------------------------
fz = clean_raw.copy().pick_channels(['EEG-Fz'])

# --- (3) PSD con API moderna ------------------------------------------
sfreq = fz.info['sfreq']
psd_obj = fz.compute_psd(method='welch',
                         fmin=1, fmax=40,
                         n_fft=int(2*sfreq),
                         n_overlap=int(1*sfreq))
psd   = psd_obj.get_data()[0]   # shape (n_freqs,)
freqs = psd_obj.freqs
psd_db = 10 * np.log10(psd)

# --- (4) detrend 1/f ---------------------------------------------------
mask = (freqs > 4) & (freqs < 40)
m, b = np.polyfit(np.log10(freqs[mask]), psd_db[mask], 1)
psd_flat = psd_db - (m*np.log10(freqs) + b)

# --- (5) potenza relativa ---------------------------------------------
total = psd_db[(freqs >= 1) & (freqs <= 40)].sum()
theta = psd_db[(freqs >= 4) & (freqs <= 7)].sum() / total
alpha = psd_db[(freqs >= 8) & (freqs <= 12)].sum() / total
print(f"Theta 4-7 Hz:  {theta*100:.1f}%\nAlpha 8-12 Hz: {alpha*100:.1f}%")

# --- (6) grafici -------------------------------------------------------
plt.figure(); plt.semilogy(freqs, psd_db); plt.title("Fz – PSD")
plt.figure(); plt.plot(freqs, psd_flat);   plt.title("Fz – Flattened PSD")
plt.show()
```

```python
# ---------------------------------------------------------------
# Analisi bande di potenza su Fz — richiede:
# • clean_raw   (dopo ICA)
# • fz          = clean_raw.copy().pick_channels(['EEG-Fz'])
# ---------------------------------------------------------------

import numpy as np
import pandas as pd
import mne

# ── 1. verifica di avere il canale Fz pulito ───────────────────
try:
    fz
except NameError:
    raise RuntimeError(
        "Variabile `fz` non trovata. "
        "Esegui prima le celle di preprocessing (lettura file, filtri, ICA) "
        "e crea `fz = clean_raw.copy().pick_channels(['EEG-Fz'])`."
    )

sfreq = fz.info['sfreq']

# ── 2. PSD 1-45 Hz con Welch ───────────────────────────────────
psd_obj = fz.compute_psd(
    method='welch', fmin=1, fmax=45,
    n_fft=int(2 * sfreq),          # finestre di 2 s
    n_overlap=int(1 * sfreq)       # overlap 50 %
)
psd, freqs = psd_obj.get_data(return_freqs=True)
psd = psd[0]                       # shape (n_freqs,)
psd_db = 10 * np.log10(psd)

# ── 3. definizione bande ───────────────────────────────────────
bands = {
    "delta 1–4 Hz":  (1, 4),
    "theta 4–7 Hz":  (4, 7),
    "alpha 8–12 Hz": (8, 12),
    "beta 13–30 Hz": (13, 30),
    "gamma 30–45 Hz": (30, 45),
}

total_power = psd_db[(freqs >= 1) & (freqs <= 45)].sum()

rows = []
for name, (fmin, fmax) in bands.items():
    mask = (freqs >= fmin) & (freqs <= fmax)
    band_power = psd_db[mask].sum()
    rows.append({
        "Band": name.split()[0],
        "fmin_Hz": fmin,
        "fmax_Hz": fmax,
        "Power_dBsum": band_power,
        "Relative_%": 100 * band_power / total_power,
    })

df = pd.DataFrame(rows)
display(df)              # mostra la tabella interattiva in Jupyter
```
